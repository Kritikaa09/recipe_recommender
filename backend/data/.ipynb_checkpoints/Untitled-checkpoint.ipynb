```python
import pandas as pd
import re
import ast
```

```python
# 1️⃣ Load dataset
DATA_PATH = "backend/data/recipes.csv"
df = pd.read_csv(DATA_PATH)
print("🔹 Original shape:", df.shape)
print("🔹 Columns:", df.columns.tolist())
```

```python
# 2️⃣ Drop rows with missing essential fields
df = df.dropna(subset=['recipe_name', 'ingredients', 'directions'])
```

```python
# 3️⃣ Convert JSON-like strings to lists if needed
def safe_eval(x):
    try:
        return ast.literal_eval(x) if isinstance(x, str) else x
    except:
        return []

df['ingredients'] = df['ingredients'].apply(safe_eval)
df['directions'] = df['directions'].apply(safe_eval)
```

```python
# 4️⃣ Clean text: lowercase + remove symbols
def clean_text_list(lst):
    clean_items = []
    for item in lst:
        text = str(item).lower()
        text = re.sub(r'[^a-zA-Z\s]', '', text)
        text = re.sub(r'\s+', ' ', text).strip()
        clean_items.append(text)
    return clean_items

df['clean_ingredients'] = df['ingredients'].apply(clean_text_list)
df['clean_directions'] = df['directions'].apply(clean_text_list)
```

```python
# 5️⃣ Merge text for embeddings
df['combined_text'] = df.apply(
    lambda row: f"{row['recipe_name'].lower()} " +
                ' '.join(row['clean_ingredients']) + ' ' +
                ' '.join(row['clean_directions']),
    axis=1
)
```

```python
# 6️⃣ Handle missing numeric fields
numeric_cols = ['prep_time', 'cook_time', 'total_time', 'servings', 'rating']
for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)
```

```python
# 7️⃣ Fill missing categorical fields
df['cuisine_path'] = df.get('cuisine_path', '').fillna('Unknown')
```

```python
# 8️⃣ Save cleaned dataset
CLEAN_PATH = "backend/data/recipes_cleaned.csv"
df.to_csv(CLEAN_PATH, index=False)
print("✅ Cleaned dataset saved as:", CLEAN_PATH)
print("✅ Cleaned shape:", df.shape)
print(df[['recipe_name', 'clean_ingredients', 'clean_directions']].head(5))
```
